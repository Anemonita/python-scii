## Gráfica de Pastel

Muestra los porcentajes de ocupación de las categorías en un eje categórico o la relación peso/etiqueta.

Consiste en una lista de pesos (uno para cada categoría) y opcionalmente una lista de nombres de cada categoría.

> Documentación: https://matplotlib.org/gallery/pie_and_polar_charts/pie_features.html#sphx-glr-gallery-pie-and-polar-charts-pie-features-py

```python
import matplotlib.pyplot as plt

# Lista de etiquetas para cada categoría
frutas = ["manzana", "mango", "fresa"]
# Lista de pesos para cada categoría
cantidades = [10, 40, 25]

# Sintaxis: plt.pie(pesos, labels=etiquetas)
plt.pie(cantidades, labels=frutas)

# Mostramos la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt

etiquetas = ["hombres", "mujeres"]
pesos = [10, 40]

plt.pie(pesos, labels=etiquetas)

plt.show()
```

Opcionalmente podemos separar las rebanadas del pastel (separar las categorías del centro) a través de `explode=separadores` definiendo una lista de separaciones para cada categoría: `0` si no queremos separar `0.1` un pequeña separación, `0.2` una separación mayor, etc.

Observa también el parámetro `autopct="%.2f%%"` que establece el forma para pintar el porcentaje sobre las rebanadas del pie a dos decimales incluyendo el símbolo `%`.

```python
import matplotlib.pyplot as plt

etiquetas = ["hombres", "mujeres"]
pesos = [10, 40]

# Creamos un separador por cada categoría
separadores = [0.2, 0]

plt.pie(pesos, labels=etiquetas,
        explode=separadores, autopct="%.2f%%")

plt.show()
```

# Gráfica de barras

Muestra las alturas/pesos de las etiquetas específicadas. Podríamos construir barras dónde los pesos equivalgan a las veces que aparece una categoría sobre el eje. Podríamos construir los pesos a partir de la suma o el promedio (o cualquier otro estadísitico) extraído otro eje escalar. Podríamos mostrar la relación peso/etiqueta.

> Documentación: https://matplotlib.org/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py


```python
import matplotlib.pyplot as plt

etiquetas = ["Manzana", "Pera", "Mango"]
pesos = [400, 100, 350]

plt.bar(etiquetas, pesos)

plt.show()
```

## Gráfica de relación (scatter)

Una gráfica de relación nos permite visualizar datos sobre dos ejes escalares a través de puntos coordenados según el eje `X` y el eje `Y`. Es útil para visualizar la dispersión de los datos y ver si hay patrones de aislamiento para formar grupos o islas de datos.

> Documentación: https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
import seaborn as sns

temperaturas = [30, 31.2, 33.8, 45, 32.17, 29.6, 28.4, 30.1, 20.4]
ph = [5, 5.1, 5.4, 7, 5.3, 4.9, 3.2, 5.05, 2.9]

sns.set() # preajusta a seaborn (opcional)

sns.scatterplot(temperaturas, ph)
```

## Gráfica de Heatmap

Muestra una matriz de datos numéricos de forma visual a través de una escala de colores. Cada elemento de la matriz tendrá un color dentro del `heatmap` y podremos visualizar los datos de la matriz sin confindirnos con los números, de ahí podrémos intuir patrones de comportamiento, por ejemplo las filas de color más alto/bajo y las columnas con colores más alto/bajo.

```python
import seaborn as sns; sns.set()

mat = [
    #X  Y  Z  W
    [1, 2, 3, 4], # A
    [5, 6, 7, 8], # B
    [2, 1, 0, 9], # C
]

sns.heatmap(mat)
```

Se puede definir la paleta de colores, según los colores de `matplotlib`. Vea https://matplotlib.org/examples/color/colormaps_reference.html

```python
import seaborn as sns; sns.set()

mat = [
    #X  Y  Z  W
    [1, 2, 3, 4], # A
    [5, 6, 7, 8], # B
    [2, 1, 0, 9], # C
]

sns.heatmap(mat, cmap="YlGnBu") # cmap="magma"
```
